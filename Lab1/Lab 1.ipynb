## Nyquist theorem: v_sample > 2 v_signal
2048 is how many samples we have so 2048 samples aka numbers
when doing fourier transform, need to have same numbers aka 2048 so need 1024 frequencies ==> np.fft gives complex fourier transform
np.fft.rfft gives you the real fft so it would only give 1024 instead of 2048 becuz 2048 is 1024 in positive and 1024 in negative side so they're mirroring.
np.fftfreq gives the frequency so wide things become skinny and skinny things become wide in frequencies. 
fourier transform has nyquist theorem built into it

resolution:  spacing between each bin aka delta v_min is 1/period and delta v_max is tthe bandwidth so it's 1/delta t (t stands for time)


## Convolution/correlation
convolution is the inverse fourier tranform of fourier transform of two functions multiplied together and correlation is the conjugate product of one of the functions of the same thing above

plotting power spectra is the absolute value of the fourier transform all squared

```python
def axis(x, y, xlabel, ylabel, title, xlim = (None, None), ylim = (None, None)):
    fig, ax = plt.subplots()
    ax.plot(x,y)
    ax.set_xlabel(f'{xlabel}', fontsize = 12)
    ax.set_ylabel(f'{ylabel}', fontsize = 12)
    ax.set_title(f'{title}', fontsize = 12)
    ax.grid(True)
    
    if not(xlim[0] == None and xlim[1] == None):
        ax.set_xlim(xlim[0], xlim[1])
    if not(ylim[0] == None and ylim[1] == None):
        ax.set_ylim(ylim[0], ylim[1])

        
        
def plot(freq, samplesize, file, v_time = False, v_freq = False, power_spectrum = False, 
         strList = False, shift = False, x_lim = False, y_lim = False):
    if strList == True:
        data = []
        with open("%s" %file) as datFile:
            data.append ([data.split() for data in datFile])
    #print (dat [0][0]) # first block of 3d array
        data_block = data[0][0] #first block of 3d array
        volt = []
        for i in range(len(data_block)):
            volt.append(float(data_block[i])) #turns the data to a float
    else:
        df = pd.read_csv('%s' %file, names=['volts'])
        volt = df["volts"]
                         
    timescale = len(volt)/(samplesize) # in ms if kHz samp frequency
    times = [i * timescale for i in range(len(volt))]
                         
    ffv = np.fft.fft(volt)
    frequency = np.fft.fftfreq(len(volt), 1/samplesize) #1/(freq))
    power = (abs(ffv))**2
                         
    if v_time == True:
        #volts v time:
        vtime_title = f'voltage time series for: \n{samplesize} kHz sample, {freq} kHz frequency'
        axis(times, volt, "time [ms]", "voltage [V]", vtime_title, xlim = (0,1000))
    if v_freq == True:
        #voltage v frequency:
        vfreq_title = f'voltage vs. frequency plot for: \n{samplesize} kHz sample, {freq} kHz frequency'
        axis(frequency, ffv, "frequency [kHz]", "voltage [V]", vfreq_title)
    
    if power_spectrum == True:
        #power spectrum
        ptitle = f'power spectrum for: \n{samplesize} kHz sample, {freq} kHz frequency'
        if shift == True:
            if x_lim == True:
                xlim = (1000, 1500)
            else:
                xlim = (None, None)
                         
            if y_lim == True:
                ylim = (0, 2000)
            else: 
                ylim = (None, None)
            axis(np.fft.fftshift(frequency), power, "frequency [kHz]", "power [$V{^2}$]", ptitle, xlim = x_lim, ylim = y_lim)
                         
                    

                    
```
